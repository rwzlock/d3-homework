```python
#Dependencies
import pandas as pd
```

```python
#Dataset location
path = "Data\data.csv"
```

```python
#create and display df
data_df = pd.read_csv(path)
data_df.head()
```

```python
#Create lists for the columns we want to keep
state = list(data_df['state'])
abbr = list(data_df['abbr'])
smokes = list(data_df['smokes'])
obesity = list(data_df['obesity'])
cleaned_data_dict = {
    "state": state,
    "abbr": abbr,
    "smokes": smokes,
    "obesity": obesity
}
#Debug
#print(len(state))
#print(state[1])
#print(len(abbr))
#print(abbr[1])
#print(len(smokes))
#print(smokes[1])
#print(len(obesity))
#print(obesity[1])
#print(cleaned_data_dict)
```

```python
#Display and create Dataframe
cleaned_data_df = pd.DataFrame(cleaned_data_dict, columns=['state', 'abbr', 'smokes', 'obesity'])
cleaned_data_df.head()
```

```python
#Export to csv
cleaned_data_df.to_csv("Data\cleaned_data.csv")
```
